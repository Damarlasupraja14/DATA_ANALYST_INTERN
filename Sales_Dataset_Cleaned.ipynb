```python
group1 = df.groupby(["Product", "Year"])["Profit"].sum().reset_index()

group2 = df.groupby(["Product"])["Order_Quantity"].mean().reset_index()

final = pd.merge(group1, group2, on="Product")
```

      here reset_index() is used because, both groups will have index while grouping so they will override with index while merging.
      In order to overcome this I have used reset_index(), it will reset the index.

## Checking and Handling Outliers

```python
columns = ["Unit_Cost", "Unit_Price", "Profit", "Cost", "Revenue"]


def handle_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower_bound, lower_bound,
                  np.where(df[column] > upper_bound, upper_bound, df[column]))
    return df
```

```python
for col in columns:
    df=handle_outliers(df,col)
```

       Outliers are a datapoint or group of datapoints that will be different from remaining datapoints,  those are away      from  the remaining datapoints.
 
        I have used capping method means, replacing the values which are less than lower_bound with lower_bound value and values which are higher than upper_bound with upper_bound
